**PASO 4** Defina el modelo del sistema y los objetivos de la simulación

El objetivo de la simulación es determinar la probabilidad de que el tiempo que una persona espere a la otra sea como máximo 10. Para lograr esto, se realizarán simulaciones para generar los tiempos de llegada de las dos personas y calcular el tiempo de espera en cada simulación. Luego, se analizarán los resultados de las simulaciones para determinar la proporción de veces en las que el tiempo de espera es menor o igual a 10, y esta proporción representará la probabilidad buscada.

**PASO 5** Genere números aleatorios.

```python
import numpy as np
import random #Librería para generar número aleatoríos
import matplotlib.pyplot as plt #Librería para gráficar
```

```python
def tiempos():
  persona1=random.randrange(0, 30, 1) #Se generan números aleatorios entre 0 y 30 para cada persona que representan su tiempo de llegada
  persona2=random.randrange(0, 30, 1)
  return(abs(persona1-persona2)) #Se obtiene la diferencia entre esos tiempos
```

**PASO 6** Diseñe el experimento

```python
def encuentro(n):
  e=0
  Lista=[]
  for i in range(n):
    if tiempos()<=10: #Si la diferencia es menor o igual a 10, se suma 1 al contador de encuentro
      e=e+1
      Lista.append(e/(i+1)) #En la variable Lista se guarda la probabilidad del encuentro
  return(Lista)
```

```python
x=encuentro(100)
```

```python
plt.plot(x) #Se gráfican las distintas probabilidades del encuentro
```

```python
np.mean(x) #Se obtiene su media
```

**PASO 7** Repita el experimento n veces

```python
def modelo(n): #Replicamos el modelo varias veces
    lista=[]
    for i in range(n):
        s = encuentro(100)
        lista.append(np.mean(s)) #Se obtiene la media de cada modelo 
    return lista
```

```python
Y=modelo(40)
```

```python
print(np.mean(Y), np.std(Y))
```

```python
plt.hist(Y) #Gráficamos las medias de los diferentes modelos
plt.show()
```

**Paso 8** Hallamos el intervalo de confianza.


```python
print("El intervalo de confianza es: ",
np.mean(Y)-(1.96)*((np.var(Y))/np.sqrt(40)), "≤ m ≤",
np.mean(Y)+(1.96)*((np.var(Y))/np.sqrt(40)))
```
